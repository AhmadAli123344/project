```python
from keras.models import load_model
model =load_model('/content/drive/MyDrive')
```

```python
model.summary()
```

```python
import os
import cv2
import pickle
import numpy as np

folder_path=["/content/drive/MyDrive/project_data/test/Acne and Rosacea",
            #  "/content/drive/MyDrive/project_data/test/Actinic Keratosis Basal Cell Carcinoma and other Malignant Lesions",
             "/content/drive/MyDrive/project_data/test/Atopic Dermatitis"
            #  "/content/drive/MyDrive/project_data/test/Bullous Disease",
            #  "/content/drive/MyDrive/project_data/test/Cellulitis Impetigo and other Bacterial Infections"
             ]
dataset=[]
y_label=[]
for i in folder_path:
  disease_name=os.path.basename(i)
  for j in os.listdir(i):
    image_path=os.path.join(i,j)
    if os.path.isfile(image_path):
            image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
            if image is not None:
                resized_image = cv2.resize(image, (300,300))
                dataset.append(resized_image/255)
                y_label.append(disease_name)

X=np.array(dataset)
```

```python
X.shape
```

```python
y_preds = model.predict(X)
```

```python
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
y_test=lab.fit_transform(y_label)
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
y_preds_argmax = np.argmax(y_preds ,axis = 1)
acc=accuracy_score(y_test,y_preds_argmax)
cf=confusion_matrix(y_test ,y_preds_argmax)
```
